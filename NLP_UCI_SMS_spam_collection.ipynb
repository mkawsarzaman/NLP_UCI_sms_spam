```python
def message_processing(message):
    nopunc = [char for char in message if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    return [word for word in nopunc.split() if word not in stopwords.words('english')]
```

```python
messages['message'].head(5).apply(message_processing)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer=message_processing)
```

```python
bow_transformer.fit(messages['message'])
print(len(bow_transformer.vocabulary_))
```

```python
message_bow = bow_transformer.transform(messages['message'])
```

```python
message_bow.shape
```

```python
message_bow.nnz
```

```python
sparsity = (100.0 * message_bow.nnz / (message_bow.shape[0] * message_bow.shape[1]))
print('sparsity: {}'.format((sparsity)))
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf = TfidfTransformer()
```

```python
message_tfidf = tfidf.fit_transform(message_bow)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(message_tfidf,messages['label'])
```

```python
predictions = model.predict(message_tfidf)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(messages['label'],predictions))
```

```python
from sklearn.model_selection import train_test_split
```

```python
message_train, message_test, label_train, label_test = train_test_split(messages['message'], messages['label'], test_size=0.3)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('bow_transformer',CountVectorizer(analyzer=message_processing)),
    ('ttfidf',TfidfTransformer()),
    ('classifier',MultinomialNB())
])
```

```python
pipeline.fit(message_train,label_train)
```

```python
predictions = pipeline.predict(message_test)
```

```python
print(classification_report(label_test,predictions))
```
